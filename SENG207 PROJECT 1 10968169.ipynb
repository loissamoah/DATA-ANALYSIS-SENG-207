# Getting  rid of all null values and duplicates.

#### Before beginning the analysis, I pre-processed the data to get rid of null values and duplicates. I checked for

#### missing values using the `isnull()` method and removed them using the `dropna()` method. I also checked for 

#### duplicates using the `duplicated()` method and removed them using the `drop_duplicates()` method.



```python
# check for missing values
print(df.isnull().sum())
```

#### In the event of null values,The `dropna()` function drops all rows with missing 

####  values (i.e. NaN values) from the data set. Setting the `inplace` parameter to  

#### `True` modifies the data set directly rather than creating a copy.


```python
# Dropping all rows with null values
df.dropna(inplace=True)

# Dropping all duplicates
df.drop_duplicates(inplace=True)

# Reset the index after dropping rows
df.reset_index(drop=True, inplace=True)

```
